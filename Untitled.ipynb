```python
modi = pd.read_csv("modi_reviews.csv")
rahul = pd.read_csv('rahul_reviews.csv')
```

```python
#drop the unwanated colun
modi.drop(['Unnamed: 0'], axis=1,inplace=True)
```

```python
modi['Tweet'] = modi['Tweet'].astype(str)
rahul['Tweet'] = rahul['Tweet'].astype(str)
```

```python
def find_polarity1(review):
    return TextBlob(review).sentiment
```

```python
find_polarity1("India is very bad country")
```

```python
def find_polarity(review):
    return TextBlob(review).sentiment.polarity
```

```python
find_polarity("India is very bad country")
```

```python
modi['Polarity'] = modi['Tweet'].apply(find_polarity)
rahul['Polarity'] = rahul['Tweet'].apply(find_polarity)
```

```python
modi
```

```python
rahul
```

```python
modi['Label'] = np.where(modi['Polarity']>0,'positive','negative')
modi['Label'][modi['Polarity']==0]='Neutral'

rahul['Label'] = np.where(rahul['Polarity']>0,'positive','negative')
rahul['Label'][rahul['Polarity']==0]='Neutral'
```

```python
# Remove rows with 'Neutral' sentiment from 'modi'
modi = modi[modi['Label'] != 'Neutral']

# Remove rows with 'Neutral' sentiment from 'rahul'
rahul = rahul[rahul['Label'] != 'Neutral']
```

```python
modi
```

```python
rahul
```

```python
modi_count=modi.groupby('Label').count()
modi_count
```

```python
rahul_count=rahul.groupby('Label').count()
rahul_count
```

```python
import matplotlib.pyplot as plt

# Assuming you have already executed the code to create 'Label' column in 'modi' and 'rahul' dataframes

# Count the occurrences of each label in 'modi'
modi_label_counts = modi['Label'].value_counts()

# Count the occurrences of each label in 'rahul'
rahul_label_counts = rahul['Label'].value_counts()

# Plotting for 'modi'
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
modi_label_counts.plot(kind='bar', color=['green', 'red', 'blue'])
plt.title('Modi Sentiment Distribution')
plt.xlabel('Sentiment Label')
plt.ylabel('Count')

# Plotting for 'rahul'
plt.subplot(1, 2, 2)
rahul_label_counts.plot(kind='bar', color=['green', 'red', 'blue'])
plt.title('Rahul Sentiment Distribution')
plt.xlabel('Sentiment Label')
plt.ylabel('Count')

plt.tight_layout()
plt.show()

```
